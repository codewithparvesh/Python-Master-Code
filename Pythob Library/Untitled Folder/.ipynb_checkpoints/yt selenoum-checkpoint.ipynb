```python
import numpy as np
day = []
with open("D:\Coding\Listed Company Assignment\Links_data.csv") as file_obj :
    data = csv.DictReader(file_obj,skipinitialspace=True)
    
    
    for row in data:
        day.append(row['Links'])
```

```python
def fun(url):
    if "hashtag" not in str(url):
        
        driver.get(url)
#         time.sleep(2) # use explicit wait
        try:
            element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, "//*[@id='owner']//a")))

            link = driver.find_elements(By.XPATH, "//*[@id='owner']//a")
    #         print(len(link))
            x = link[0].get_attribute('href')
            if x:
                chnl_lst.append(x)
            else:
                chnl_lst.append(url)
        except:
            pass
```

```python
chnl_list=[]
```

```python
def fun(url):
    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        bigboxes = soup.find("div")
        a = bigboxes.find_all("link")[1]
        chnl_list.append(a['href'])
    except:
        pass
```

```python
cnt=1
for i in uni_day:
    print(len(chnl_list),"  cnt",cnt)
    cnt+=1
    if cnt%10==0:
        dict = {'Links': chnl_list}  
        df = pd.DataFrame(dict) 
        df.to_csv('Links_data.csv') 
    fun(i)
```
