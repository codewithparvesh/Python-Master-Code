```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
driver = webdriver.Chrome(executable_path ="chromedriver_win32\chromedriver.exe")

# Navigate to Google
driver.get("https://www.google.com/")
```

```python

import time
# Scrape the links from the search results
links = []
cnt = 0
cnt_2 = 0
while len(links) < 10000:
#     print("While loop","     ",cnt_2)
    # Find all the search result links
#     search_results = driver.find_elements(By.CSS_SELECTOR,'.yuRUbf > a')
    search_results = driver.find_elements(By.TAG_NAME,'a')
    for result in search_results:
        link = result.get_attribute("href")
        print(len(links))
#         print("for loop","       ",cnt)
        cnt+=1
        if link not in links:
            links.append(link)
        if len(links) == 10000:
            break

# Print the links
for link in links:
    print(link)
```
