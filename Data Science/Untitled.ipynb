```python
import numpy as np
import csv

with open(r"C:\Users\dc\Downloads\terrorismData.csv") as file_obj :
    data = csv.DictReader(file_obj,skipinitialspace=True)
    df = list(data)


    
ans =[]
for row in df:
    ans.append([row['Day'], row['Month'],row['Year']])

ans=np.array(ans)
cnt=0
for row in ans:
    if int(float(row[0])) > 0 and int(float(row[1]))==1 and int(float(row[2]))==2010:
        cnt+=1
print(cnt)
```
